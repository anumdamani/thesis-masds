```python
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

# Get the texts for bestsellers and non-bestsellers
df = pd.read_csv("/Users/anumdamani/thesis/fiction_combined_21025/updated_fiction_official.csv", dtype={'year': 'Int64'})
bestseller_texts = df[df['label'] == 1]['title']
non_bestseller_texts = df[df['label'] == 0]['title']

# Function to create the word cloud
def create_wordcloud(texts, title, ax):
    # TF-IDF Vectorization
    vectorizer = TfidfVectorizer(max_features=50, stop_words='english')
    X = vectorizer.fit_transform(texts)
    
    # Get the words and their corresponding TF-IDF scores
    tfidf_scores = dict(zip(vectorizer.get_feature_names_out(), X.sum(axis=0).A1))
    
    # Create the WordCloud
    wordcloud = WordCloud(background_color="white", width=800, height=400, max_words=50).generate_from_frequencies(tfidf_scores)
    
    # Plot the word cloud
    ax.imshow(wordcloud, interpolation='bilinear')
    ax.axis("off")
    ax.set_title(title)

# Plot WordClouds for Bestsellers and Non-Bestsellers
fig, ax = plt.subplots(1, 2, figsize=(14, 7))

create_wordcloud(bestseller_texts, "Word Cloud of Important Words in Bestselling Book Titles", ax[0])
create_wordcloud(non_bestseller_texts, "Word Cloud of Important Words in Non-Bestselling Book Titles", ax[1])

# Save both word clouds separately
fig.savefig('/Users/anumdamani/thesis/EDA/wordcloud_bestsellers_and_nonbestsellers_titles.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Apply VADER sentiment analysis to book descriptions
def get_vader_sentiment(text):
    sentiment = analyzer.polarity_scores(str(text))
    return sentiment

bestsellers['vader_sentiment'] = bestsellers['description'].apply(get_vader_sentiment)
nonbestsellers['vader_sentiment'] = nonbestsellers['description'].apply(get_vader_sentiment)

# Check the sentiment scores
print(bestsellers[['title', 'vader_sentiment']].head())
print(nonbestsellers[['title', 'vader_sentiment']].head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Function to get the compound sentiment score
def get_compound_sentiment(text):
    return sia.polarity_scores(text)['compound']

# Apply the function to the description
bestsellers['compound_sentiment'] = bestsellers['description'].apply(get_compound_sentiment)
nonbestsellers['compound_sentiment'] = nonbestsellers['description'].apply(get_compound_sentiment)
# Calculate the average compound sentiment for each group
avg_bestseller_sentiment = bestsellers['compound_sentiment'].mean()
avg_nonbestseller_sentiment = nonbestsellers['compound_sentiment'].mean()

# Create a DataFrame for plotting
sentiment_data = pd.DataFrame({
    'Group': ['Bestsellers', 'Non-Bestsellers'],
    'Average Sentiment': [avg_bestseller_sentiment, avg_nonbestseller_sentiment]
})
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns

# Define paths to image folders
bestseller_folder = "/Users/anumdamani/thesis/fiction_images/1_bestseller_covers"
nonbestseller_folder = "/Users/anumdamani/thesis/fiction_images/0_nonbestseller_covers"

# Function to compute RGB histograms efficiently
def compute_rgb_histogram(image_folder, num_bins=50, sample_size=100):
    red_vals, green_vals, blue_vals = [], [], []
    image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

    # Randomly sample images to avoid memory overload
    sample_files = np.random.choice(image_files, min(len(image_files), sample_size), replace=False)

    for img_name in sample_files:
        img_path = os.path.join(image_folder, img_name)
        image = Image.open(img_path).convert("RGB")
        image = image.resize((150, 150))  # Resize to save memory
        image_array = np.array(image)

        # Flatten and store RGB values
        red_vals.extend(image_array[:, :, 0].flatten())
        green_vals.extend(image_array[:, :, 1].flatten())
        blue_vals.extend(image_array[:, :, 2].flatten())

    # Compute histograms
    red_hist, _ = np.histogram(red_vals, bins=num_bins, range=(0, 255))
    green_hist, _ = np.histogram(green_vals, bins=num_bins, range=(0, 255))
    blue_hist, _ = np.histogram(blue_vals, bins=num_bins, range=(0, 255))

    return red_hist, green_hist, blue_hist

# Compute histograms for bestsellers and non-bestsellers
red_b, green_b, blue_b = compute_rgb_histogram(bestseller_folder)
red_nb, green_nb, blue_nb = compute_rgb_histogram(nonbestseller_folder)

# Plot histograms
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
colors = ['red', 'green', 'blue']
titles = ['Red Channel', 'Green Channel', 'Blue Channel']

for i, (bestseller_hist, nonbestseller_hist, color, title) in enumerate(zip(
        [red_b, green_b, blue_b], [red_nb, green_nb, blue_nb], colors, titles)):

    axes[i].bar(range(50), bestseller_hist, color=color, alpha=0.6, label="Bestsellers")
    axes[i].bar(range(50), nonbestseller_hist, color=color, alpha=0.3, label="Non-Bestsellers")
    axes[i].set_title(title)
    axes[i].set_xlabel("Pixel Intensity Bins")
    axes[i].set_ylabel("Frequency")
    axes[i].legend()

plt.suptitle("RGB Color Channel Distributions for Bestsellers vs. Non-Bestsellers")
plt.tight_layout()
#plt.show()
plt.savefig('/Users/anumdamani/thesis/EDA/color_histograms.png', dpi=300, bbox_inches='tight')
```
